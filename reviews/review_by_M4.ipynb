Now I was able to run the full analysis and view the scatter plot and line plot, I could tackle the question on wether the correct answers follow a hidden deliberate pattern.

The scatterplot shows the mean answer value for each 100 questions.It shows that the mean answer values follow a pattern, they are very close to whole numbers 1.0, 2.0, 3.0 and 4.0. While the mean technically shows what the majority chose and not the correct answer it, the consistency of the mean values shows that the 64 respondents successfully identified the answer pattern.
The plot also shows to have a wave like pattern, this signifies a deliberate arrangement of answers. 

The line plot shows every respondent's answers for each question alltogether. 
Alot of the lines overlap at whole numbers such as 1.0, 2.0, 3.0 and 4.0 with quite a few downward spikes at 0 (representing any unanswered questions). I noticed a red line at 0.0 from question 21 to 100, meaning that the respondent stopped the quiz early. These overlaps show that multiple people selected the same answer. As many of these lines follow the same zigzag pattern it suggests that there is a deliberate pattern of answers.

Overall, based on both plots there is clear evidence of a deliberate pattern of a correct answer sequence. So my integration script successfully managed to import modules, extract the relevant information and analyse the data to reveal thhis structure.
